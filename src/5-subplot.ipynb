```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12, 10))

# Subplot 1: Line Chart
plt.subplot(2, 2, 1)
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title('Line Chart')
plt.grid(True)

# Subplot 2: Bar Chart
plt.subplot(2, 2, 2)
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 2, 5]
plt.bar(categories, values, color=['red', 'green', 'blue', 'orange'])
plt.title('Bar Chart')

# Subplot 3: Scatter Plot
plt.subplot(2, 2, 3)
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 3, 5, 7, 11]
plt.scatter(x_scatter, y_scatter, color='purple')
plt.title('Scatter Plot')
plt.grid(True)

# Subplot 4: Pie Chart
plt.subplot(2, 2, 4)
sizes = [30, 25, 20, 15, 10]
labels = ['A', 'B', 'C', 'D', 'E']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')

plt.tight_layout()
plt.show()
```
